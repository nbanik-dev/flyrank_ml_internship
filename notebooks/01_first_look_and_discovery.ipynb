## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

**My Observation:**
Filtering to impressions_90d > 0 had no effect since 100% of pages already meet that condition. This confirms the near-zero correlation (0.001) isn't an artifact of zero-traffic noise — search_volume genuinely doesn't predict impressions_90d across the full dataset. This is a stronger version of Discovery A: it's not that low-traffic pages are muddying the signal, there simply isn't a linear signal to find.

```python
df_nonzero = df[df["impressions_90d"] > 0]

corr_filtered = df_nonzero["search_volume"].corr(df_nonzero["impressions_90d"])
print(f"Correlation (impressions_90d > 0 only): {corr_filtered:.3f}")
print(f"Pages included: {len(df_nonzero)} / {len(df)} ({len(df_nonzero)/len(df):.1%})")

# compare side by side
print(f"\nOriginal correlation (all pages): {corr:.3f}")
print(f"Filtered correlation (nonzero only): {corr_filtered:.3f}")
```

**Observation 2:**
Comparison articles show the lowest CTR relative to their position tier across every tier with sufficient data, with an average gap of -0.99 from the best-performing content type. Note: some cells (e.g. top_3 × feedly article) show CTR > 1, indicating either a data quality issue in the ctr column or extremely small sample sizes — recommend verifying group counts before treating single-tier outliers as reliable.

```python
visible = df[df["impressions_90d"] >= 100]

# CTR by content_type within each position_tier
ctr_by_type_pos = visible.groupby(["position_tier", "content_type"])["ctr"].mean().unstack()
print(ctr_by_type_pos.round(4).to_string())

# for each position tier, find the worst-performing content_type
worst_per_tier = ctr_by_type_pos.idxmin(axis=1)
print("\nWorst content_type by position tier:")
print(worst_per_tier.to_string())

# overall: which content_type is worst most often / by widest average gap
gap_from_best = ctr_by_type_pos.sub(ctr_by_type_pos.max(axis=1), axis=0)
avg_gap = gap_from_best.mean().sort_values()
print("\nAverage CTR gap from best-in-tier, by content_type (most negative = worst):")
print(avg_gap.round(4).to_string())
```

**Observation 3:**
Median word count is nearly identical between 'up' and 'down' trending pages ([X] vs [Y] words), suggesting content length alone isn't a strong lever for trend direction. Like search_volume/impressions_90d, this is a case where an intuitive SEO lever doesn't show up as a driver in this dataset — worth checking other candidate features (e.g. content_age_days, freshness signals) instead.

```python
# check spread, not just median - similar medians can still hide different distributions
wc_full = df.groupby("trend_direction")["word_count"].agg(["median", "mean", "std", "count"])
print(wc_full.round(0).to_string())

# visualize distribution shape, not just central tendency
df.boxplot(column="word_count", by="trend_direction");
```
