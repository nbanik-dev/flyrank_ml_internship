## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**Decision:** Which pages should an SEO content editor review first this week, out of thousands with real search visibility?

**Who acts, and how:** A content editor with limited time (say, 20-50 reviews/week) uses a ranked queue with reason codes to decide where to spend that time — refresh, expand, protect, or leave alone.

**Cost of a wrong call:** A false positive wastes an editor's review hours on a page that wasn't actually at risk. A false negative means a real decliner with meaningful traffic gets missed until it's worse. Given editor time is the scarce resource, precision in the top of the queue (precision@K) matters more than catching every possible decliner.

**Why ML, not a plain rule:** A hand-written rule (e.g. "flag pages with falling position AND low freshness") only checks a few signals at a time and can't learn how signals interact — that's exactly what notebook 01 showed: the rule reached ~24% precision@50, a model reached ~74%. The pattern is real but too tangled for an if-statement.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

declining = (df["trend_direction"] == "down").sum()
print(f"Pages currently trending down: {declining} / {len(df)} ({declining/len(df):.1%})")

visible = df[df["impressions_90d"] >= 100]
declining_visible = visible[visible["trend_direction"] == "down"]
print(f"Visible pages (impressions_90d >= 100): {len(visible)} ({len(visible)/len(df):.1%} of all)")
print(f"Declining AND visible: {len(declining_visible)} ({len(declining_visible)/len(visible):.1%} of visible)")

med = visible.groupby("trend_direction")["impressions_90d"].median()
print("\nMedian impressions_90d by trend, visible pages only:")
print(med.round(0).to_string())
```
